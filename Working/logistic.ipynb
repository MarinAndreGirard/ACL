```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic map function
def logistic_map(x, r):
    return r * x * (1 - x)

# Parameters
r_values = np.linspace(2.5, 4.0, 10000)  # Range of r
iterations = 1000  # Number of iterations
last = 100  # Plot only the last 100 iterations for stability

# Initialize the figure
plt.figure(figsize=(10, 7))

# Loop over r values
for r in r_values:
    x = 0.5  # Initial condition
    # Iterate the logistic map
    for i in range(iterations):
        x = logistic_map(x, r)
        # Plot the last 100 values for each r
        if i >= (iterations - last):
            plt.plot(r, x, ',k', alpha=0.25)

# Labeling the plot
plt.title("Bifurcation diagram of the logistic map")
plt.xlabel("r")
plt.ylabel("x")
plt.xlim(2.5, 4)
plt.show()
```

The bifurcation diagram at right summarizes this. The horizontal axis shows the possible values of the parameter r while the vertical axis shows the set of values of x visited asymptotically from almost all initial conditions by the iterates of the logistic equation with that r value.



```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic map function
def logistic_map(r, x):
    return r * x * (1 - x)

# Set up parameters
iterations = 1000
last = 100  # How many points at the end of iterations to plot
r_values = np.linspace(2.5, 4.0, 10000)
x_initial = 1e-5  # Initial population

# Prepare the plot
plt.figure(figsize=(10, 7))

# Loop over each r value
for r in r_values:
    x = x_initial
    # Iterate to get rid of transient behavior
    for _ in range(iterations - last):
        x = logistic_map(r, x)
    
    # Plot the last points
    x_vals = np.zeros(last)
    for i in range(last):
        x = logistic_map(r, x)
        x_vals[i] = x
    
    plt.plot([r] * last, x_vals, ',k', alpha=0.25)

# Add labels and show the plot
plt.title('Bifurcation Diagram of the Logistic Map')
plt.xlabel('r')
plt.ylabel('x')
plt.show()
```
