```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

def draw_arrows(ax, arrows, labels, color='black'):
    for (start, end), label in zip(arrows, labels):
        ax.annotate('', xy=end, xytext=start,
                    arrowprops=dict(facecolor=color, edgecolor=color, arrowstyle='->'))
        mid_point = ((start[0] + end[0]) / 2, (start[1] + end[1]) / 2)
        ax.text(mid_point[0], mid_point[1], label, fontsize=12, ha='center')
# Define colors
upper_color = 'lightblue'
lower_color = 'lightgreen'
arrow_colors = ['blue', 'red', 'green', 'orange']

# Function to draw colored blocks
def draw_blocks(ax, upper_color, lower_color):
    ax.add_patch(plt.Rectangle((-1, 0), 2, 1, color=upper_color, alpha=0.5))
    ax.add_patch(plt.Rectangle((-1, -1), 2, 1, color=lower_color, alpha=0.5))

def draw_blocks_mixing(ax, upper_color, lower_color):
    ax.add_patch(plt.Rectangle((-1, -1), 1, 1, color=lower_color, alpha=0.5))
    ax.add_patch(plt.Rectangle((0, 0), 1, 1, color=lower_color, alpha=0.5))
    ax.add_patch(plt.Rectangle((-1, 0), 1, 1, color=upper_color, alpha=0.5))
    ax.add_patch(plt.Rectangle((0, -1), 1, 1, color=upper_color, alpha=0.5))




# Define colors for the vectors
vector_colors = ['blue', 'red', 'green', 'orange']

# Function to draw arrows from the origin
def draw_vectors(ax, vectors, labels, middle_labels, colors):
    for vector, label, mid_label, color in zip(vectors, labels, middle_labels, colors):
        ax.annotate('', xy=vector, xytext=(0, 0),
                    arrowprops=dict(facecolor=color, edgecolor=color, arrowstyle='->', lw=2))
        ax.text(vector[0]*1.1, vector[1]*1.1, label, fontsize=12, ha='center', va='center', color=color)
        mid_point = (vector[0] / 2+0.2, vector[1] / 2)
        ax.text(mid_point[0], mid_point[1], mid_label, fontsize=12, ha='center', va='center', color=color)


def draw_curved_arrows(ax, curved_arrows, color='black'):
    for start, end, label in curved_arrows:
        # First arrow with a positive curve
        arrow1 = patches.FancyArrowPatch(start, end, connectionstyle="arc3,rad=0.3",
                                         arrowstyle='-|>', color=color, lw=1.5,
                                         mutation_scale=20)  # Adjust mutation_scale for arrowhead size
        ax.add_patch(arrow1)
        mid_point1 = ((start[0] + end[0]) / 2, (start[1] + end[1]) / 2)
        ax.text(mid_point1[0], mid_point1[1], label, fontsize=12, ha='center', va='center', color=color)
        
        # Second arrow with a negative curve
        arrow2 = patches.FancyArrowPatch(end, start, connectionstyle="arc3,rad=0.3",
                                         arrowstyle='-|>', color=color, lw=1.5,
                                         mutation_scale=20)  # Adjust mutation_scale for arrowhead size
        ax.add_patch(arrow2)
        mid_point2 = ((start[0] + end[0]) / 2, (start[1] + end[1]) / 2)
        

# Vectors for the states
vectors = [(0.8, 0.8), (0.8, -0.8), (-0.8, 0.8), (-0.8, -0.8)]
labels = ['|E2>', '|E4>', '|E1>', '|E3>']
middle_labels=['|12>', '|22>', '|11>', '|21>']
middle_labels_internal=['|11>', '|22>', '|12>', '|21>']
middle_labels_mixing=['|22>','|12>' , '|11>', '|21>']
empty_labels = ['', '', '', '']
# Curved arrows for relabeling
curved_arrows_internal = [((0.8, 0.8), (0.8, -0.8), '12'), ((-0.8, 0.8), (-0.8, -0.8), '22')]
curved_arrows_mixing = [((0.8, 0.8), (-0.8, 0.8), '12'), ((0.8, -0.8), (-0.8, -0.8), '21')]

# Updated color configuration
vector_colors_basic = ['blue', 'blue', 'blue', 'blue']
vector_colors_basic2 = ['blue', 'green', 'blue','green']
vector_colors_mixing = ['green', 'blue', 'blue','green']


# Draw the figure with the updated requirements
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Top Left: Basic transitions
axs[0, 0].set_title("Hilbert space")
draw_vectors(axs[0, 0], vectors, labels,empty_labels, vector_colors_basic)

# Top Right: Split Transitions
axs[0, 1].set_title("With labels")
draw_blocks(axs[0, 1], upper_color, lower_color)
draw_vectors(axs[0, 1], vectors, labels,middle_labels, vector_colors_basic2)

# Bottom Left: Internal relabeling
axs[1, 0].set_title("Internal relabeling")
draw_blocks(axs[1, 0], upper_color, lower_color)
draw_vectors(axs[1, 0], vectors, labels,middle_labels_internal, vector_colors_basic2)
draw_curved_arrows(axs[1, 0], [((0.8, 0.8), (-0.8, 0.8), ''), ((0.8, 0.8), (-0.8, 0.8), '')], color='black')


# Bottom Right: Mixing relabeling
axs[1, 1].set_title("Mixing relabeling")
draw_blocks_mixing(axs[1, 1], upper_color, lower_color)
draw_vectors(axs[1, 1], vectors, labels,middle_labels_mixing, vector_colors_mixing)
draw_curved_arrows(axs[1, 1], [((0.8, 0.8), (0.8, -0.8), ''), ((0.8, 0.8), (0.8, -0.8), '')], color='black')

# Adjust layout
for ax in axs.flat:
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
